```python
import numpy as np
from matplotlib import pyplot as plt
import random
import scipy.linalg as spla

def dg(A):
    return A.conj().T

random.seed(1)

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

```python
# Model parameters.
γ = 1
δ = 0
Ω = 1
dΩ = 1e-4

# Initial state.
ψ0 = np.array([1, 0], dtype="complex128")

# Simulation parameters.
number_trajectories = 100
t_final = 50
dt = 0.01
```

```python
# Operators.
σ = np.array([[0,1],[0,0]], dtype="complex128")
c = np.sqrt(γ) * σ
σz = np.array([[1,0], [0,-1]], dtype="complex128")

# Hamiltonian and jump operator.
H = - δ * dg(σ) @ σ + Ω/2. * (dg(σ) + σ)
M1 =  np.sqrt(γ) * σ

# Displaced operators to compute the derivatives.
Hp = - δ * dg(σ) @ σ + (Ω + dΩ)/2. * (dg(σ) + σ)
Hm = - δ * dg(σ) @ σ + (Ω - dΩ)/2. * (dg(σ) + σ)
M1p = M1
M1m = M1

# The jump operators have to be passed as a list.
M_l = [M1]
Mp_l = [M1p]
Mm_l = [M1m]
```

```python
def gillespie_fisher(H, Hp, Hm, M_l, Mp_l, Mm_l, dθ, ψ0, t_final, dt, number_trajectories, verbose=True):
    print("Gillespie-Fisher method - Welcome")
    print("=> Starting the pre-computation stage")
    t_range = np.arange(0, t_final, dt)
    
    # Constructs the overall jump operator.
    J = np.zeros_like(M_l[0])
    for M in M_l:
        J += dg(M) @ M
    # Effective (non-Hermitian) Hamiltonian.
    He = H - 1j/2. * J
    
    # Constructs the no-jump evolution operators for all the relevant times.
    V = [] # List of the no-jump evolution operators.
    Qs = [] # List of the non-state-dependent part of the waiting time distribution.
    for t in t_range:
        ev_op = spla.expm(-1j * He * t)
        V.append(ev_op)
        nsd_wtd = dg(ev_op) @ J @ ev_op
        Qs.append(nsd_wtd)
    # Prints the matrix norm of the latest Qs.
    error = spla.norm(Qs[-1])
    print(f"-> Truncation error given by norm of latest Qs matrix -> {error}")
    
    # Displaced versions of the effective Hamiltonian.
    Jp = np.zeros_like(Mp_l[0])
    for Mp in Mp_l:
        Jp += dg(Mp) @ Mp
    Hep = Hp - 1j/2. * Jp
    Jm = np.zeros_like(Mm_l[0])
    for Mm in Mm_l:
        Jm += dg(Mm) @ Mm
    Hem = Hm - 1j/2. * Jm
    
    # Vector of the derivatives of the evolution operator wrt the parameter.
    Vdot = []
    for t in t_range:
        vd = (spla.expm(-1j * Hep * t) - spla.expm(-1j * Hem * t)) / (2. * dθ)
        Vdot.append(vd)
    # Derivatives of all the jump operators.
    Mdot = []
    for n_M in range(len(M_l)):
        Mdot.append((Mp_l[n_M] - Mm_l[n_M]) / (2 * dθ))
    # Precomputation of all the derivatives of MkV(t).
    Δ = []
    for n_M in range(len(M_l)):
        fixed_M = []
        for n_t, t in enumerate(t_range):
            obj = Mdot[n_M] @ V[n_t] + M_l[n_M] @ Vdot[n_t]
            fixed_M.append(obj)
        Δ.append(fixed_M)
    
    # Evolution cycle.
    print("=> Starting the evolution cycle.")
    
    # List for the results.
    trajectories_results = []
    
    # Cycle over the trajectories.
    for trajectory in range(number_trajectories):
        print(f"-> Starting trajectory {trajectory}")
        
        # Initial state.
        ψ = ψ0
        # Absolute time.
        τ = 0
        # Initial ξ matrix.
        ξ = np.zeros((2,2), dtype="complex128")
        
        results = []
        dict_initial = {"AbsTime": 0,
                       "TimeSinceLast": 0,
                       "JumpChannel": None,
                       "ξAfter": ξ,
                       "ψAfter": ψ0}
        results.append(dict_initial)
        
        while τ < t_final:
            dict_jump = {}
            
            # Pass to density matrix formalism.
            ρ = np.outer(dg(ψ), ψ)
            
            # Compute the waiting time distribution, exploiting the pre-computed part.
            Ps = []
            for Q in Qs:
                # Even though the number is real, theree might be small imaginary noise.
                wtd = np.real(dg(ψ) @ Q @ ψ)
                Ps.append(wtd)
                
            # Sample from the waiting time distribution.
            n_T = random.choices(range(len(t_range)), weights=Ps)[0]
            # TODO: what happens if T brings us beyond t_final?

            # Increase the absolute time.
            τ += t_range[n_T]
            dict_jump["AbsTime"] = τ
            dict_jump["TimeSinceLast"] = t_range[n_T]

            # Update the state.
            ψ = V[n_T] @ ψ
            # Choose where to jump.
            weights = []
            for M in M_l:
                weight = np.real(dg(ψ) @ dg(M) @ M @ ψ)
                weights.append(weight)
            n_jump = random.choices(range(len(M_l)), weights=weights)[0]
            dict_jump["JumpChannel"] = n_jump
            # Update the state after the jump.
            ψ = M_l[n_jump] @ ψ
            norm = spla.norm(ψ)
            # Renormalize the state.
            ψ = ψ / norm
            
            if verbose:
                print(dict_jump)
                        
            # Compute the ξ matrix.
            ξ = 1./(norm**2) * (M_l[n_jump] @ V[n_T] @ ξ @ dg(V[n_T]) @ dg(M_l[n_jump]))
            ξ += 1./(norm**2) * (Δ[n_jump][n_T] @ ρ @ dg(V[n_T]) @ dg(M_l[n_jump]))
            ξ += 1./(norm**2) * (M_l[n_jump] @ V[n_T] @ ρ @ dg(Δ[n_jump][n_T]))
            dict_jump["ξAfter"] = ξ
            dict_jump["ψAfter"] = ψ
            dict_jump["trξ2"] = (np.real(np.trace(ξ)))**2
        
            results.append(dict_jump)
                
        trajectories_results.append(results)
    
    return trajectories_results, V, Vdot, t_range
```

```python
def fisher_at_time_on_trajectory(t_range, relevant_times, V, Vdot, trajectory_data, ψ0):
    v_fisher = []
    
    # Creates an array of jump times.
    jump_times = [trajectory_data[i]["AbsTime"] for i in range(len(trajectory_data))]
    # Creates an array of states after the jumps.
    ψ_after_jumps = [trajectory_data[i]["ψAfter"] for i in range(len(trajectory_data))]
    # Creates an array of ξ after the jumps.
    ξ_after_jumps = [trajectory_data[i]["ξAfter"] for i in range(len(trajectory_data))]
        
    # Cycles over the jumps times.
    for n_jump, t_jump in enumerate(jump_times[:-1]):
        next_jump_time = jump_times[n_jump + 1]
        # Determines the set of relevant times between this jump and the following one.
        relevant_times_in_interval = [t for t in t_range if t_jump <= t < next_jump_time]
        # Initial state and monitoring operator immediately after the jump.
        # Cycles over the relevant times.
        for t_abs in relevant_times_in_interval:
            ψ = ψ_after_jumps[n_jump]
            ρ = np.outer(dg(ψ), ψ) 
            ξ = ξ_after_jumps[n_jump]
            # This time is the closest one for which we have available a pre-stored
            # version of the V(t) evolution operators.
            n_t = find_nearest(t_range, t_abs - t_jump)
            trace = np.real(np.trace(V[n_t] @ ρ @ dg(V[n_t])))
            ξ = Vdot[n_t] @ ρ @ dg(V[n_t]) + V[n_t] @ ρ @ dg(Vdot[n_t]) + V[n_t] @ ξ @ dg(V[n_t])
            ξ = ξ / trace
            fisher = np.real((np.trace(ξ))**2)
            v_fisher.append(fisher)
            
    # Now computes the ξ for all times after the latest jump.
    last_jump_absolute_time = jump_times[-1]
    relevant_times_after_last_jump = [t for t in t_range if t >= last_jump_absolute_time]
    for t_abs in relevant_times_after_last_jump:
        ψ = ψ_after_jumps[-1]
        ρ = np.outer(dg(ψ), ψ)
        ξ = ξ_after_jumps[-1]
        # This time is the closest one for which we have available a pre-stored
        # version of the V(t) evolution operators.
        n_t = find_nearest(t_range, t_abs - last_jump_absolute_time)
        trace = np.real(np.trace(V[n_t] @ ρ @ dg(V[n_t])))
        ξ = Vdot[n_t] @ ρ @ dg(V[n_t]) + V[n_t] @ ρ @ dg(Vdot[n_t]) + V[n_t] @ ξ @ dg(V[n_t])
        ξ = ξ / trace
        fisher = (np.trace(ξ))**2
        v_fisher.append(fisher)
    
    return v_fisher
```

```python
trajectories_results, V, Vdot, t_range = gillespie_fisher(H, Hp, Hm, M_l, Mp_l, Mm_l, dΩ, ψ0, t_final, dt, number_trajectories, verbose=False)

average_fisher = np.zeros(len(t_range))
for n_trajectory, trajectory in enumerate(trajectories_results):
    print(f"Now filling the gaps for trajectory {n_trajectory}")
    v_fisher = fisher_at_time_on_trajectory(t_range, t_range, V, Vdot, trajectory, ψ0)
    v_fisher = np.array(v_fisher)
    average_fisher += v_fisher / number_trajectories
    plt.plot(t_range, v_fisher)
    
plt.show()
plt.plot(t_range, average_fisher)
plt.show()
```

```python
# Check the waiting time distribution, to see if Gillespie works.
v_waitings = []
n_traj = 100
trajectories_results, V, Vdot, t_range = gillespie_fisher(H, Hp, Hm, M_l, Mp_l, Mm_l, dΩ, ψ0, t_final, dt, n_traj, False)
for trajectory in trajectories_results:
    v_waitings += [trajectory[i]["TimeSinceLast"] for i in range(len(trajectory))]
    
v_waitings_clean = [w for w in v_waitings if w != 0]
plt.hist(v_waitings_clean, bins=50)
```
